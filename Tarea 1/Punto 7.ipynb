Considerar la ecuacion diferencial:

$$
 u(t)'= \alpha u(t)
$$

El metodo de euler consiste en aproximar la solucion de la ecuacion diferencial de la siguiente forma:

$$
u_{n+1}= u_n + u'_n \Delta t
$$

Remplazando $u'(t)$:
$$
u_{n+1}= u_n + \alpha u_n \Delta t
$$

Reescribiendo la ecuacion anterior:

$$
u_{n}= u_{n-1} + \alpha u_{n-1} \Delta t
$$

$$
u_{n}= u_{n-1} (1+ \alpha \Delta t)
$$

Siguiendo con el metodo iterativo:

$$
u_{n-1}= u_{n-2} (1+ \alpha \Delta t)
$$

Remplazando la ecuacion para $u_{n-1}$ en la ec. para $u_n$

$$
u_{n}= u_{n-2} (1+ \alpha \Delta t) (1+ \alpha \Delta t)= (1+ \alpha \Delta t)^2 u_{n-2}
$$

Repetir el proceso $n$ veces: 

$$
u_{n-2}= u_{n-3} (1+ \alpha \Delta t)
$$

$$
u_{n-3}= u_{n-4} (1+ \alpha \Delta t)
$$

$$
\vdots
$$

$$
u_{1}= u_{0} (1+ \alpha \Delta t)
$$

Se remplazan las ecuaciones de cada $u_i$ en su respectivo $u_{i+1}$.Notese que el termino $(1+\alpha \Delta t)$ aparece $n$ veces, por lo tanto:

$$
u_{n}= u_{0} (1+ \alpha \Delta t)^n
$$


Las soluciones oscilaran si el termino $(1+ \alpha \Delta t)<0$, ya que  $(1+ \alpha \Delta t)^n>0$ para $n$ par, y $(1+ \alpha \Delta t)^n<0$ para $n$ impar.

En el caso que $\alpha<0$:

$$
(1+ \alpha \Delta t)<0
$$

$$
(\frac{1}{\alpha}+ \Delta t)<0
$$

$$
\Delta t > \frac{-1}{\alpha}
$$








```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solucion_num(alfa,deltat,u0,tmax):
    
    t_=np.arange(0,tmax,deltat)
    
    u_=np.zeros_like(t_)
    
    for k in range(len(u_)):
        
        u_[k]=(1+ alfa*deltat)**k *u0
    
    return (t_,u_)
    
    
    
    
```

```python
def solucion_exac(t,alfa, u0):
    
    return u0*np.exp(alfa*t)
    
    
```

```python
alfa=-1
tmax=20
u0=1000
t=np.linspace(0,tmax,100)
```

```python

fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot()
ax.plot(solucion_num(alfa,1.1,u0,tmax)[0],solucion_num(alfa,1.1,u0,tmax)[1],label="$\Delta T = 1.1 s$")
ax.plot(solucion_num(alfa,1.5,u0,tmax)[0],solucion_num(alfa,1.5,u0,tmax)[1],label="$\Delta T = 1.5 s$")
ax.plot(solucion_num(alfa,1.9,u0,tmax)[0],solucion_num(alfa,1.9,u0,tmax)[1],label="$\Delta T = 1.9 s$")
ax.plot(t,solucion_exac(t,alfa,u0),label="Sol.exacta")
ax.legend()
```
