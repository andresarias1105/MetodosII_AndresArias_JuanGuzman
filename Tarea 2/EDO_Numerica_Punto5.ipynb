```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(v,u,q):
    
    return u**q
```

```python
def solexacta(t,q):
    if q == 1:
        u = np.exp(t)
    else:
        u = ((1-q)*t+1)**(1/(1-q))
        
    return u
```

```python
def GetRungeKutta4(x,y,h,f,q):

    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1],q)
        k2 = f(x[i-1]+0.5*h,y[i-1]+0.5*h*k1,q)
        k3 = f(x[i-1]+0.5*h,y[i-1]+0.5*h*k2,q)
        k4 = f(x[i-1]+h,y[i-1]+h*k3,q)
        
        y[i]= y[i-1]+ h/6 *(k1+2*k2+2*k3+k4)
```

```python
N = 1000
t = np.linspace(0,10,N)
h = t[1] - t[0]
```

```python
q = [0,0.2,0.4,0.7,0.9,1]
y_sol = []

for i in range(len(q)):
    y_ = np.zeros_like(t)
        
    if q[i] == 1:
        y_[0] = 1
        
    else:
        y_[0] = 1
        
    GetRungeKutta4(t,y_,h,f,q[i])
        
    y_sol.append(y_)
```

```python
y_ex = []
for i in range(len(q)):
    y_ex.append(solexacta(t,q[i]))
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 5))

ax[0][0].plot(t,y_sol[0],label="Sol.Numerica".format(q[0]))
ax[0][0].plot(t,y_ex[0],label="Sol.Exacta".format(q[0]))
ax[0][0].set_title("q = 0")
ax[0][0].legend()

ax[0][1].plot(t,y_sol[1],label="Sol.Numerica".format(q[1]))
ax[0][1].plot(t,y_ex[1],label="Sol.Exacta".format(q[1]))
ax[0][1].set_title("q = 0.2")
ax[0][1].legend()

ax[0][2].plot(t,y_sol[2],label="Sol.Numerica".format(q[2]))
ax[0][2].plot(t,y_ex[2],label="Sol.Exacta".format(q[2]))
ax[0][2].set_title("q = 0.4")
ax[0][2].legend()

ax[1][0].plot(t,y_sol[3],label="Sol.Numerica".format(q[3]))
ax[1][0].plot(t,y_ex[3],label="Sol.Exacta".format(q[3]))
ax[1][0].set_title("q = 0.7")
ax[1][0].legend()

ax[1][1].plot(t,y_sol[4],label="Sol.Numerica".format(q[4]))
ax[1][1].plot(t,y_ex[4],label="Sol.Exacta".format(q[4]))
ax[1][1].set_title("q = 0.9")
ax[1][1].legend()

ax[1][2].plot(t,y_sol[5],label="Sol.Numerica".format(q[5]))
ax[1][2].plot(t,y_ex[5],label="Sol.Exacta".format(q[5]))
ax[1][2].set_title("q = 1")
ax[1][2].legend()

plt.tight_layout()
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 5))

ax[0][0].plot(t,y_ex[0]-y_sol[0],label="Sol.Numerica".format(q[0]))
ax[0][0].set_title("Error para q = 0")
ax[0][0].legend()

ax[0][1].plot(t,y_ex[1]-y_sol[1],label="Sol.Numerica".format(q[1]))
ax[0][1].set_title("Error para q = 0.2")
ax[0][1].legend()

ax[0][2].plot(t,y_ex[2]-y_sol[2],label="Sol.Numerica".format(q[2]))
ax[0][2].set_title("Error para q = 0.4")
ax[0][2].legend()

ax[1][0].plot(t,y_ex[3]-y_sol[3],label="Sol.Numerica".format(q[3]))
ax[1][0].set_title("Error para q = 0.7")
ax[1][0].legend()

ax[1][1].plot(t,y_ex[4]-y_sol[4],label="Sol.Numerica".format(q[4]))
ax[1][1].set_title("Error para q = 0.9")
ax[1][1].legend()

ax[1][2].plot(t,y_ex[5]-y_sol[5],label="Sol.Numerica".format(q[5]))
ax[1][2].set_title("Error para q = 1")
ax[1][2].legend()

plt.tight_layout()
```

En las gráficas se puede notar un crecimiento en el orden de magnitud del error
