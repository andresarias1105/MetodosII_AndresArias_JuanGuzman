(a) Muestre que el error del método está descrito por:

\begin{equation}
    \epsilon_{n+1} - (2+h^{2}a^{'}_n)\epsilon_{n} +\epsilon_{n-1}=0
\end{equation}
 
En primer lugar, es necesario partir de la fórmula iterativa original de la posición para luego sumarle un error asociado a cada valor del pasado, presente y futuro:

\begin{equation}
    \vec{r}_{n+1} = 2\vec{r}_n -\vec{r}_{n-1} + \vec{a}(\vec{r}_n)h^{2}
\end{equation}

\begin{equation}
    \vec{r}_{n+1} +\epsilon_{n+1} = 2\vec{r}_n + 2 \epsilon_{n} -\vec{r}_{n-1} - \epsilon_{n-1} + \vec{a}(\vec{r}_n + \epsilon_n)h^{2}
\end{equation}

Por Taylor, es posible decir que $\vec{a}(\vec{r}_n + \epsilon_n)$ es aproximadamente $\vec{a}(\vec{r}_n) + \epsilon_n\vec{a}^{'}(\vec{r}_n)$ porque $\epsilon_n$ es un valor pequeño. Además, para cambiar la notación, se escribe $\vec{a}(\vec{r}_n)$ como $\vec{a}_n$, por lo que se tiene que:

\begin{equation}
    \vec{r}_{n+1} +\epsilon_{n+1} = 2\vec{r}_n + 2 \epsilon_{n} -\vec{r}_{n-1} - \epsilon_{n-1} + (\vec{a}_n + \epsilon_n\vec{a}^{'}_n)h^{2}
\end{equation}

\begin{equation}
    \vec{r}_{n+1} +\epsilon_{n+1} = 2\vec{r}_n + 2 \epsilon_{n} -\vec{r}_{n-1} - \epsilon_{n-1} + \vec{a}_{n}h^{2}+ \epsilon_{n}\vec{a}^{'}_{n}h^{2}
\end{equation}

Ahora, se separan los errores de la ecuación anterior para obtener lo siguiente:

\begin{equation}
    \epsilon_{n+1} = 2 \epsilon_{n} - \epsilon_{n-1} + \epsilon_{n}a^{'}_{n}h^{2}
\end{equation}

Finalmente, la expresión queda como:

\begin{equation}
    \epsilon_{n+1} - (2+h^{2}a^{'}_{n})\epsilon_{n} + \epsilon_{n-1} = 0
\end{equation}


(b) Para el caso de un oscilador armónico clásico muestre que:

\begin{equation}
    \epsilon_{n+1} - 2(1-R)\epsilon_{n} +\epsilon_{n-1}=0
\end{equation}

donde $2R = h^{2}\omega^{2}$


Tenemos que $a_x = -\omega^2x$ para un oscilador armónico clásico, por lo que calculando la derivada con respecto a x se obtiene que:

\begin{equation}
    \frac{da_x}{dx} = -\omega^2
\end{equation}

Reemplazando en la expresión del error se llega a:

\begin{equation}
    \epsilon_{n+1} - (2-h^{2}\omega^2)\epsilon_{n} + \epsilon_{n-1} = 0
\end{equation}

\begin{equation}
    \epsilon_{n+1} - 2(1-\frac{h^{2}\omega^2}{2})\epsilon_{n} + \epsilon_{n-1} = 0
\end{equation}

Finalmente, definiendo $2R = h^{2}\omega^{2}$:

\begin{equation}
    \epsilon_{n+1} - 2(1-R))\epsilon_{n} + \epsilon_{n-1} = 0
\end{equation}

(c) Usando la suposición de función potencia $ \epsilon_n = \epsilon_{0}\lambda^{n}$, muestre que las raíces son:

\begin{equation}
    \lambda_{\pm} = 1 + R \pm \sqrt{R^2 -2R}
\end{equation}

Reemplazando la igualdad en la fórmula para un oscilador armónico clásico y operando los términos nos da que:

\begin{equation}
    \epsilon_0 \lambda^{n+1} - (2-2R)\epsilon_0\lambda^{n} + \epsilon_{0}\lambda^{n-1}=0
\end{equation}

\begin{equation}
    \lambda^{n+1} - (2-2R)\lambda^{n} + \lambda^{n-1}=0
\end{equation}

\begin{equation}
    \lambda - (2-2R) + \lambda^{-1}=0
\end{equation}

\begin{equation}
    \frac{\lambda^{2}+1}{\lambda} - (2-2R)=0
\end{equation}

\begin{equation}
    \lambda^{2} - (2-2R)\lambda + 1= 0
\end{equation}

Nos queda un polinomio de grado 2, por lo que se puede usar la fórmula cuadrática para obtener lo siguiente:

\begin{equation}
    \lambda_{\pm} = \frac{(2-R) \pm \sqrt{(2-2R)^2 - 4}}{2}
\end{equation}

\begin{equation}
    \lambda_{\pm} = \frac{(2-2R) \pm \sqrt{4R^{2}-8R+4 - 4}}{2}
\end{equation}

\begin{equation}
    \lambda_{\pm} = \frac{(2-2R) \pm \sqrt{4R^{2}-8R}}{2}
\end{equation}

Finalmente:

\begin{equation}
    \lambda_{\pm} = 1 - R \pm \sqrt{R^{2}-2R}
\end{equation}

(d) El valor de la estabilidad se obtiene para $|\lambda_{\pm}| = 1$, Muestre que el paso de integración debe cumplir:

\begin{equation}
    h \leq \frac{2}{\omega}
\end{equation}

Tenemos que $\lambda_{\pm} = \pm1$, así que reemplazando en la expresión del punto anterior:

\begin{equation}
    \pm1 = 1 - R \pm \sqrt{R^{2}-2R}
\end{equation}

\begin{equation}
    \pm1- 1 + R = \pm \sqrt{R^{2}-2R}
\end{equation}

Elevendo al cuadrado ambos lados:

\begin{equation}
    \pm1- 1 + R = R^{2}-2R
\end{equation}

\begin{equation}
    2 - (\pm2) \pm 2R - 2R + R^2 = R^{2}-2R
\end{equation}

\begin{equation}
    2 - (\pm2) \pm 2R = 0
\end{equation}

\begin{equation}
    \pm 2R = (\pm2)- 2
\end{equation}

Reemplazando ese valor en $\pm 2R =  \pm h^2 \omega^{2}$:

\begin{equation}
   (\pm2)- 2 = \pm h^2 \omega^{2}
\end{equation}

\begin{equation}
  h^2 = \pm \frac{(\pm2)- 2}{\omega^{2}}
\end{equation}

Como el paso se define como una cantidad del conjunto de los números reales mayores que cero, queda como:

\begin{equation}
  h^2 = -\frac{(-2)- 2}{\omega^{2}}
\end{equation}

\begin{equation}
  h^2 = \frac{4}{\omega^{2}}
\end{equation}

\begin{equation}
  h = \frac{2}{\omega}
\end{equation}

En este caso, $h = \frac{2}{\omega}$ sería un valor crítico necesario para la estabilidad del algoritmo, por lo que un paso menor o igual a ese sería el esencial, así que:

\begin{equation}
  h \leq \frac{2}{\omega}
\end{equation}

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
from tqdm import tqdm 
```

```python
def oscilador(x,omega):
    return -omega**2*x
```

```python
def Verlet(a,x0,v0,t,omega,h):
    
    
    
    x=np.zeros_like(t)
    v=np.zeros_like(t)
    
    x[0]=x0
    x[1]=x[0] + h*v0
    v[0]=v0
    v[1]=x[0] + h*a(x[1],omega)
    
    for i in range(2,len(x)):
        
        x[i]= 2*x[i-1]-x[i-2]+ a(x[i-1],omega)*h**2
        v[i] = (x[i]-x[i-2])/(2*h)
        
    return x,v
    
    
    
    
    
```

# Estable $h<2/\omega$

```python
omega=np.pi
h=0.1 #estable
t=np.arange(0,10,h)

x0=1
v0=1
```

```python
x_,vx_= Verlet(oscilador,x0,v0,t,omega,h)
```

```python
scale=1
t1= t[::scale]
X_=x_[::scale]
VX=vx_[::scale]

fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(121)
ax1 = fig.add_subplot(122)

def init():
    

    ax.clear()
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)

    

    ax1.clear()
    ax1.set_ylim(-2,2)
   
    
def Update(i):
    
    init()
    
    x = X_[i]
    
  
    

    vx =VX[i]
    #ax.quiver([0,x],0,vx,length=0.5)
    
    circle = plt.Circle( (0,x), 0.3, color='r', fill=False )
    ax.add_patch(circle)
       
    ax.arrow(0,x,0,vx,color='k',head_width=0.1,length_includes_head=True)
    
    ax1.plot( t1[0:i] , X_[0:i],color="r")
       
    
        
    
Animation = anim.FuncAnimation(fig,Update,frames=len(t1),init_func=init)
```
