```python
import numpy as np
import matplotlib.pyplot as plt
from  tqdm import tqdm
```

```python
def formato_grafica(titulo, ejex, ejey, leyenda=False, xlim=[None, None], ylim=[None, None]):
    plt.rcParams['axes.axisbelow'] = True

    plt.title(titulo, fontsize=15)
    plt.ylabel(ejey, fontsize=13)
    plt.xlabel(ejex, fontsize=13)

    plt.tick_params(direction='out', length=5, width=0.75, grid_alpha=0.3)
    plt.xticks(rotation=0)
    plt.minorticks_on()
    plt.ylim(ylim[0], ylim[1])
    plt.xlim(xlim[0], xlim[1])
    plt.grid(True)
    plt.grid(visible=True, which='major', color='grey', linestyle='-')
    plt.grid(visible=True, which='minor', color='lightgrey', linestyle='-', alpha=0.2)

    if leyenda == True:
        plt.legend(loc='best')

    plt.tight_layout;
```

```python
x = np.linspace(-5,5,1000)

w = 1
m = 1
h_barra = 1
```

```python
def V(x):
    return 0.5*m*w**2*x**2
```

```python
def Numerov(x,E,V):
    R = (2*m*V(x)/h_barra**2) - (2*m*E/h_barra**2)
    S = np.zeros_like(R)
    
    psi = np.zeros_like(x)
    psi[0] = 0
    psi[1] = 1e-5
    
    h = x[1]-x[0]
    
    for i in range(1,len(psi)-1):
        
        a = h**2*(S[i+1]+10*S[i]+S[i-1])/12
        b = 2*(1 + 5*h**2*R[i]/12)*psi[i]
        c = (1-h**2*R[i-1]/12)*psi[i-1]
        d = 1 - h**2*R[i+1]/12
        
        psi[i+1] = (a + b - c) / d
        
    return psi
```

```python
E1 = np.arange(0,6,0.5)

E_levels1 = []

for i in E1:
    if Numerov(x,i,V)[-1]*Numerov(x,i+0.001,V)[-1] < 0:
        E_levels1.append(i)
```

```python
E_levels1
```

```python
for i in range(len(E_levels1)):
    E_ = E_levels1[i]
    plt.plot(x,Numerov(x,E_,V)/np.max(Numerov(x,E_,V)),label="$E_{0} = {1}$".format(i,E_))

formato_grafica("Primeros seis estados propios para el potencial armonico","x","$\Psi(x)$",leyenda=True)
 
```

```python
x2 = np.linspace(-5,5,1000)

def Vgaussian(x):
    return -10*np.exp(-x**2/20)
```

```python
E2 = np.arange(-10,-5.5,0.001)

E_levels2 = []

for i in tqdm(E2):
    if Numerov(x2,i,Vgaussian)[-1]*Numerov(x2,i+0.001,Vgaussian)[-1] < 0 and len(E_levels2)<5 :
        E_levels2.append(i)
        
        
```

```python
E_levels2=np.round(E_levels2,3)

E_levels2
```

```python
for i in range(len(E_levels2)):
    E_2 = E_levels2[i]
    plt.plot(x,Numerov(x,E_2,Vgaussian)/np.max(Numerov(x,E_2,Vgaussian)),label="$E_{0} = {1}$".format(i,E_2))

formato_grafica("Primeros cinco estados propios para el potencial gaussiano","x","$\Psi(x)$",leyenda=True)
```

```python
x3 = np.linspace(-5,5,1000)

def Vrational(x):
    return -4/(1+x**2)**2
```

```python
E3 = np.arange(-2.5,0,0.001)

E_levels3 = []

for i in tqdm(E3):
    if Numerov(x3,i,Vrational)[-1]*Numerov(x3,i+0.001,Vrational)[-1] < 0 and len(E_levels3)<2 :
        E_levels3.append(i)
```

```python
for i in range(len(E_levels3)):
    E_3 = E_levels3[i]
    plt.plot(x,Numerov(x,E_3,Vrational)/np.max(Numerov(x,E_3,Vrational)),label="$E_{0} = {1}$".format(i,E_3))

formato_grafica("Primeros seis estados propios para el potencial racional","x","$\Psi(x)$",leyenda=True)
```
